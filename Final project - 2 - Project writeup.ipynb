### Project Problem and Hypothesis
* What's the project about? What problem are you solving?

This data set provides info to help predict behavior to retain customers. Analyzing all relevant customer data and develop focused customer retention programs.
A telecommunications company is concerned about the number of customers leaving their landline business for cable competitors. They need to understand who is leaving. 
The main problem to solve is to find out who is leaving and why. 

The hypothesis defined is: Are the customers that signed for a month contract churning more than the ones with 1 or 2 year contracts?

* Where does this seem to reside as a machine learning problem? Are you predicting some continuous number, or predicting a binary value?

It seems to reside as a machine learning problem because we have a set of defined characteristics and variables from each consumer as inputs and only one binary final outcome that will be if the user churned or not. 

* What kind of impact do you think it could have?

The impact is related entirely to revenue, the less users we churned the better the company will be and the more strategies that can be linked to prevent this situation to happen. 

* What do you think will have the most impact in predicting the value you are interested in solving for?

I think that some variables such as type of contract, amount of money to pay and number of services signed up will be the main variables impacting the final result.
